### 5. Compare the t-distribution and the normal distribution using `python`. Assume a dataset size of 6 samples.
  - For a chosen critical value $t$ e.g. 3 ($3\sigma$ rule) calculate ...
     - the probability of the two tails and
     - the probability of not being in the tails
     - for normal distribution
     - for t-distribution
  - Calculate the *actual* critical t-value assuming the probability from the two-tails of the normal distribution for that particular critical value $t$.
  - Calculate the *actual* critical z-value assuming the probability from the two-tails of the t-distribution for that particular critical value $t$.
  - What mistake could each of this correspond to and what would be the outcome?
  - Increase the degree of freedom (e.g. to 10000) what do you observe?

* Useful package: `from scipy.stats import norm, t`
* Useful functions: `t.ppf(...), t.cdf(...), norm.ppf(...), norm.cdf(...)`

```python
sample_size = 6
degrees_of_freedom = sample_size - 1
critical_value_t = 3
```

```python
prob_one_tail_normal = 1 - norm.cdf(critical_value_t)
prob_two_tails_normal = prob_one_tail_normal * 2
prob_two_tails_normal
```

```python
prob_one_tail_t = 1 - t.cdf(critical_value_t, degrees_of_freedom)
prob_two_tails_t = prob_one_tail_t * 2
prob_two_tails_t
```

```python
prob_not_in_the_tail_normal = 1 - prob_two_tails_normal
prob_not_in_the_tail_normal
```

```python
prob_not_in_the_tail_t = 1 - prob_two_tails_t
prob_not_in_the_tail_t
```

```python
alpha_norm = prob_two_tails_normal
```

```python
critical_t_value = t.ppf(1 - alpha_norm / 2, degrees_of_freedom)
critical_t_value
```

```python
alpha_t = prob_two_tails_t
```

```python
critical_z_value = norm.ppf(1 - alpha_t / 2)
critical_z_value
```

```python
new_degrees_of_freedom = 10000
```

```python
new_critical_t_value = t.ppf(1 - alpha_norm / 2, new_degrees_of_freedom)
new_critical_t_value
```

```python
new_alpha_t = 2 * (1 - t.cdf(critical_value_t, new_degrees_of_freedom))
new_critical_z_value = norm.ppf(1 - new_alpha_t / 2)
new_critical_z_value
```

5. **Answer:**

Applying normal distribution to the low number of samples of using a t-distribution for large samples. Low critical value leads to increasing the likelihood that the null-hypothesis will be rejected. It is a type 1 error. Using a larger critical value leads to the error type 2 when actually the null hypothesis is false but we do not fail it.

### 6. Apply both the z-value test and the t-value test to the dataset $\{ 3, 8, 6, 15, 13, 7 \}$ for a two-tails probability mass of 0.05.

  - Report all z-values and t-values for each sample.
  - Find the appropriate critical t-value and z-value (first manually in the table, then verify with python).
  - Apply the decision rule for both.
  - What do you observe?

```python
data = np.array([3, 8, 6, 15, 13, 7])
mean = np.mean(data)
std = np.std(data)
alpha = 0.05
df = len(data) - 1
```

```python
t_values = (data - mean) / (std / np.sqrt(len(data)))
t_values
```

```python
z_values = zscore(data)
z_values
```

```python
critical_t = t.ppf(1 - alpha / 2, df)
critical_t
```

```python
critical_z = norm.ppf(1 - alpha / 2)
critical_z
```

Outliers by t-test

```python
data[np.abs(t_values) > critical_t]
```

```python
data[np.abs(z_values) > critical_z]
```

6. **Answer:**

The t-test gives multiple outliers at the same time the z-test gives no outliers.
